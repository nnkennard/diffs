```python
import openreview, time
GUEST_CLIENT = openreview.Client(baseurl="https://api.openreview.net")
```

```python
INVITATION = "ICLR.cc/2022/Conference/-/Blind_Submission"
forum_notes = list(openreview.tools.iterget_notes(GUEST_CLIENT, invitation=INVITATION))
print(len(forum_notes), "forum notes in total")
current_time = time.time()
for i, forum_note in enumerate(forum_notes):
    if i % 10 == 0:
        new_time = time.time()
        print(f"forum {i}\t{new_time - current_time:.3f} secs")
        current_time = new_time
    notes = GUEST_CLIENT.get_all_notes(forum=forum_note.id, sort='tcdate')
    for note in notes:
        # skip the forum note
        if note.id in [forum_note.id, forum_note.original]:
            continue
        
        invitation = note.invitation.split('/')[-1]
        initiators = [s.split("/")[-1] for s in note.signatures]
        assert note.referent is None
        
#         print(invitation, "\t", initiators)
#         print("\t", note.content.keys())
        
#     raise
        
        # add known types here
        unknown = False
        if invitation == "Official_Comment":
            if initiators[0] != "Authors" and not initiators[0].startswith("Reviewer") and "Chair" not in initiators[0]:
                unknown = True
        elif invitation == "Official_Review":
            if not initiators[0].startswith("Reviewer"):
                unknown = True
        elif invitation == "Decision":
            if not "Chair" in initiators[0]:
                unknown = True
        elif invitation != "Public_Comment":
            unknown = True

        # by the end of the exploration, there should be no unknown
        if unknown:
            print(invitation, "\t", initiators)
            print("\t", note.content.keys())
            raise
```
